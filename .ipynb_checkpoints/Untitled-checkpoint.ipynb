```python
import pandas as pd
```

```python
df = pd.DataFrame(
{
    "id":["C","B","M","P"],
    "Descrip":["Capital","Bsas","Mendoza","Pampa"]
}
)
genti=["Porteño","Bonaerense","Mendocino","Pampeano"]
df=df.assign(gentilicio=genti)
df=df.assign(poblado="si")
df=df.assign(escapital="x")
df
```

```python
df["escapital"]=df["id"].apply(lambda x : "si" if x == 'C'    else "no")
df.loc[df['id'].isin(['C' , 'B']), 'poblado'] = 'no'  
df
```

```python
def cambioGentilicio (fila):
    if fila.id == "B" :
        return "Oriundo de "+fila.Descrip
    else:
        return "Oriundo de "+fila.Descrip

df["gentilicio"]=df.apply(cambioGentilicio, axis=1)
df
```

```python
dfciudad = pd.DataFrame(
{
    "idCiudad":[1,2,3,4,5],
    "NombreCiudad":["CABA","MARDEL","PINAMAR","LA PLATA","SAN RAFAEL"],
    "idProvincia":["C","B","B","B","M"]
}
)
dfciudad
```

```python
mix = pd.merge(df,dfciudad, left_on="id", right_on="idProvincia")
mix
```
